**Consegna**: Implementare in Python, all’interno del proprio Notebook, l’esempio della somma di due dadi. Usando la libreria matplotlib, crearne poi il grafico pdf.  

Viene usato il grafico di probabilità di densità (pdf) per visualizzare in maniera grafica la distribuzione dei dati; ovvero serve per rappresentare le VC (Variabili Casuali ) come istogrammi e capire gli intervalli di valori più probabili e l'andamento generale della distribuzione.

````python
import matplotlib.pyplot as plt
import numpy as np

######### DATASET #########
# Genera tutte le possibili combinazioni dei due dadi
dice = [1, 2, 3, 4, 5, 6]
combinations = [(1,1), (1,2), (1,3), (1,4), (1,5), (1,6), (2,1), (2,2), (2,3), (2,4), (2,5), (2,6), (3,1), (3,2), (3,3), (3,4), (3,5), (3,6), (4,1), (4,2), (4,3), (4,4), (4,5), (4,6), (5,1), (5,2), (5,3), (5,4), (5,5), (5,6), (6,1), (6,2), (6,3), (6,4), (6,5), (6,6)]

# Calcola le somme dei due dadi
sums = np.sum(combinations, axis=1)

# Calcola la distribuzione di probabilità delle somme
counts, bins = np.histogram(sums, bins=np.arange(1, 14))

# Normalizza i conteggi per ottenere le probabilità
probs = counts / np.sum(counts)

# Calcola la probabilità più alta
max_prob = np.max(probs)

# Crea un istogramma delle probabilità
plt.bar(bins[:-1], probs, width=0.9, align='edge')

# Aggiungi etichette e titoli al grafico
plt.xticks(bins[:-1] + 0.5, bins[:-1])
plt.xlabel('Somme dei due dadi')
plt.ylabel('Probabilità')
plt.title('Distribuzione di probabilità del lancio di due dadi')

# Mostra il grafico dei risultati ottenuti
plt.show()

# Stampa la probabilità più alta
print('Probabilità maggiore: ',max_prob)


````

**Conclusione**: la probabilità più alta ottenuta dalla distribuzione di probabilità del lancio di due dadi corrisponde a max_prob (0.166666), che rappresenta la somma di due numeri che è uguale a 7. Ciò significa che ottenere come somma 7 è l'evento più probabile quando si lanciano due dadi.

Viene richiesto di calcolare la media e la deviazione standard assieme in quanto si ottene una comprensione più completa dei dati; poichè:

  >-la *Deviazione Standard*: rappresenta una misura di dispersione o variabilità dei dati rispetto alla media. Viene usata per capire di quanto i dati si discostano dalla media.

In altre parole la media e la dev. standard ci aiutano ad identificare pattern, anomalie o differenze significative nei dati.

**Consegna**: Implementare la funzione valore assoluto su Notebook usando sia Excel che Python e calcolarne il coeff. di correlazione di Pearson

La correlazione è una misura statistica che ci serve per capire quanto i valori della variabile hanno valore predittivo rispetto ai valori della seconda variabile. Supponedo di avere due variabili 'X' e 'Y' la correlazione ci dice quanto i valori di 'X' possono predire il comportamento di 'Y'. Il coefficiente di correlazione "**r**" misura il grado in cui Y è una funzione di X (e viceversa).
Il valore di correlazione è compreso nell'intervallo [-1,1] dove:  
  >-quando assume valore "-1" significa che le variabili sono  strettamente correlate tra loro ma in maniera inversa,  
  >-quando assume vaolore "0" significa che le due variabili non sono sono correlate tra loro, ovvero sono "indipendenti",  
  >-quando assume valore "1" significa che le due variabili sono sono correlate positivamente tra loro.

In questo caso è richiesto di utilizzare Pearson per calcolare il valore di correlazoine.  
La correlazione di Pearson, chiamata anche correlazione lineare, è una misura statistica che valuta la relazione lineare tra due variabili continue fornendo un valore compreso tra -1 e 1. E' utilizzato molto nei predittori lineari nella forma Y= m*x+b.  
Però ci sono casi in cui, anche se c'è una forte correlazione lineare tra le due variabili il coefficiente di correlazione di Pearson è uguale a '0.  
Esempio: funzione valore assoluto --> Y=|x|

````python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.stats import pearsonr

# Definisci la funzione valore assoluto
def absolute_value(x):
    return abs(x)

######### DATASET #########
# Crea un array di valori x da -10 a 10 con incrementi di 0.1
x = np.arange(-3, 3.5, 0.5)

# Calcola i valori y corrispondenti alla funzione valore assoluto
y = absolute_value(x)

# Stampa i valori di x e y separati da virgola
x_str = list(map(str, x))
y_str = list(map(str, y))

x_str_with_commas = ", ".join(x_str)
y_str_with_commas = ", ".join(y_str)

print('x=',x_str_with_commas)
print('y=',y_str_with_commas)

# Crea il grafico
fig, ax = plt.subplots()
ax.plot(x, y)

# Aggiungi una linea tratteggiata orizzontale a y = 0
ax.axvline(x=0, color='black', linestyle='--')

# Aggiungi un titolo al grafico
plt.title('Funzione Valore Assoluto')

# Aggiungi un nome all'asse x
plt.xlabel('x')

# Aggiungi un nome all'asse y
plt.ylabel('y')

# Visualizza il grafico
plt.show()

# Calcolo coeff. di correlazione di Pearson utilizzando la funzione pearsonr()
pearson_correlation, p_value = pearsonr(x, y)

#Arrotondamento del coefficiente in 6 cifre decimali
pearson_correlation = round(pearson_correlation, 6)

# Pearson_correlation conterrà il valore del coefficiente di correlazione di Pearson tra x e y. Questo valore sarà compreso tra -1 e 1
print('pearson_correlation: ',pearson_correlation)

````

Il coefficiente di Spearman è un'altra statisctica che può essere usata per calcolare il valore di correlazione. E' una misura statistica non parametrica che valuta la relazione tra due variabili. A differenza della correlazione di Pearson, Spearman valuta il numero di coppie di punti in input che sono
fuori ordine e non quanto bene i punti seguimo la retta; infatti si usa molto quando bisogna lavorare con relazioni non-lineari e con outliers in quanto esso si calcola utilizzando i ranghi delle osservazioni invece dei valori effettivi delle variabili, rendendolo meno sensibile gli outlier.


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr

######### DATASET #########
temperatura = np.array([0, 5, 10, 15, 20, 30, 40, 60, 80, 100, 150, 200, 250, 300, 374])
Vi = np.array([0.0179, 0.0179, 0.0179, 0.0179, 0.0179, 0.0180, 0.0180, 0.0183, 0.0185, 0.0187, 0.0196, 0.2070, 0.0224, 0.0252, 0.0570])

# Calcola i coefficienti di correlazione di Pearson e Spearman
pearson_coef, pearson_p = pearsonr(temperatura, Vi)
spearman_coef, spearman_p = spearmanr(temperatura, Vi)

# Calcola la retta di regressione lineare
slope, intercept = np.polyfit(temperatura, Vi, 1)
line = slope * temperatura + intercept

# Disegna un grafico a dispersione, dove temperatura e Vi rappresentano le coordinate dei punti nel grafico. L'asse x rappresenta la temperatura e l'asse y rappresenta Vi.
plt.scatter(temperatura, Vi)

# Traccia una linea nel grafico. La linea è colorata di rosso, e viene etichettata come "Reg. Lineare" 
plt.plot(temperatura, line, color='red', label='Reg. Lineare')

# Aggiunge le etichette agli assi x e y e al titolo del grafico
plt.xlabel('Temperatura')
plt.ylabel('Vi')
plt.title('Grafico temperatura vs Vi')


# Aggiungi la legenda
plt.legend()
plt.show()

# Aggiungi le informazioni sui coefficienti di correlazione
x = np.float64(pearson_coef)
x = round(x, 6)
print("")

print("Il coefficiente di Pearson è: ", x)

y = np.float64(spearman_coef)
y = round(y, 6)
print("Il coefficiente di Spearman è: ", y)


````

Il coefficiente di correlazione con shif (autocorrelazione), è usata molto perchè è una correlazione che tiene conto dello spostamento o del ritardo tra due serie di dati temporali. In altre parole, misura la correlazione tra due serie temporali considerando il possibile effetto di uno spostamento tra di loro.  
Viene calcolato confrontando i valori di due serie temporali spostate l'una rispetto all'altra. Per ogni possibile spostamento, viene calcolato il coefficiente di correlazione, solitamente Pearson, tra le due serie di dati. Questo processo viene ripetuto per tutti i possibili spostamenti e selezionando il valore massimo come coefficiente di correlazione con shift.  
Un esempio di algoritmo efficiente per calolare coeff. di correlazione (con shift) è il *Fast Fourier Transform* (FFT) che permette di costruire la funzione di autocorrelazione in O(n log n)

````python
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy import signal

# Definisci la funzione valore assoluto
def absolute_value(x):
    return abs(x)

# Crea un array di valori x da 0 a 27 con incrementi di 1
x = np.arange(0, 28, 1)

# Calcola i valori y corrispondenti alla funzione traffico
traffico = [100,90,80,80,90,60,40,105,95,80,85,95,60,35,100,90,75,80,90,65,40,105,90,80,85,90,55,35]

# Calcola i valori y corrispondenti alla funzione traffico
coeffCorr = scipy.signal.correlate(x, traffico, mode='full', method='auto')

numero_spostamenti=range(1,25)
acf=[np.corrcoef(traffico[:-spostamento],traffico[spostamento:])[0,1] for spostamento in numero_spostamenti]

print(coeffCorr)
# Crea il grafico
fig, ax = plt.subplots()
ax.plot(x, traffico)

# Aggiungi titolo e label degli assi
plt.title('Traffico Milanese')
plt.xlabel('Giorni')
plt.ylabel('Traffico')

# Visualizza il grafico
plt.show()

# Crea il grafico
fig1, ax1 = plt.subplots()
ax1.plot(range(1,25), acf, linestyle='-', marker='o')
plt.axhline(0, color='grey')

# Aggiungi titolo e label degli assi
plt.title('AutoCorrelazione')
plt.xlabel('Giorni')
plt.ylabel('Correlazione')

# Visualizza il grafico
plt.show()
````

**Consegna**: Implementare in Python una funzione per il calcolo automatico del punteggio ELO negli scacchi e creare qualche simulazione di calcolo del punteggio.

**Elo-ranking** è una tecnica che serve per valutare e classificare in maniera coerente i giocatori di vari sport (scacchi, spor/giochi competitivi, ...)  
L'obiettivo è quello di assegnare a ogni singolo giocatore un punteggio (numerico), può variare nel tempo in base ai risultati delle partite giocate. Il punteggio assegnato ai giocatori rappresenta approssimativamente l'abilità relativa  del giocatore rispetto agli altri; ed una volta determinare la loro forza relativa si può calcolare le probabilità di vittoria di ciascun giocatore in un dato confronto.  
Il funzionamento consiste nel:  
  >Se un giocatore vince contro un avversario con un punteggio Elo più alto,    guadagnerà punti Elo, mentre perderà punti se viene sconfitto da un avversario con un punteggio Elo più basso.  

  >Più è alta la differenza di punteggio Elo tra i giocatori, meno punti 
   verranno guadagnati o persi in caso di vittoria o sconfitta

L’Elo-ranking di un giocatore è calcolato dalla seguente forula:
  >$r′$ (A) = $r$(A) + k($S_A$ − $E_A$)

Dove:
- $r'(A)$ rappresenta il nuovo punteggio Elo del giocatore A.
- $r(A)$ rappresenta il punteggio Elo attuale del giocatore A.
- $k$ è il fattore K, che influenza l'entità dell'aggiornamento del punteggio Elo.
- $S_A$ è il risultato ottenuto dal giocatore A nella partita (1 per la vittoria, 0 per la sconfitta, 0.5 per un pareggio).
- $E_A$ rappresenta l'aspettativa di risultato del giocatore A calcolata in base al punteggio Elo attuale di entrambi i giocatori.

      

Soluzione: la funzione `calculateNewElo` ritorna il nuovo score successivamente a una serie di giocate in cui il giocatore vince / perde contro persne che hanno un ELO differente.

````python

#Viene simulata una partita di scacchi in cui:
#ELO Avversari e vincite / perdite relative sono precalcolate
ELO_attuale = 1580
k= 32
EA = 2.353
somma = 0

#Giocate 0 = persa, 0.5 = patta, 1 = vinta
Giocate = [0,1,0.5,1,0]
ELO_avversari = [1700,1577,1388,1620,1710]

def calculateNewElo(elo_actual,somma,k,ea):
  return elo_actual + k * (somma - ea)

for partite in Giocate:
    somma+=partite

NuovoELO = calculateNewElo(ELO_attuale ,somma, k , EA)

print("L'ELO prima di effettuare partite risulta: " +str(ELO_attuale) )

for eloPoint in ELO_avversari:
    print("Il giocatore gioca contro una persona con ELO:" + str(eloPoint) )

print("L'ELO dopo le partite risulta: " +str(NuovoELO) )
````

**Conclusione**: L'Elo ranking è un sistema utilizzato per valutare le prestazioni dei giocatori in giochi competitivi. In questo caso l giocatore è partito con un punteggio Elo di 1580 e, dopo aver disputato tutte le partite previste, ha ottenuto un punteggio di 1584. Questo incremento di 4 punti Elo indica che il giocatore ha ottenuto risultati leggermente superiori alle aspettative, poiché il punteggio Elo si basa sul confronto delle prestazioni con quelle degli avversari.

###KS Test ( Kolmogorov-Smirnov KS-Test ):
Lavora sulle funzioni cumulative di due distribuzioni e determina quanto sono simili tra loro. Va a vedere la massima distanza tra le due curve, cioe vedere per ogni punto quanto le due curve differiscono.

###Test di permutazione: 
Ci dice che se una nostra ipotesi è supportata dai dati è meno probabaile che il set di dati mescolati continuino a supportare questa ipotesi.
L’aver determinato che lo stuidio delle due medie sia statisticamente significativo non deve essere vero se mischio i dati maschi / femmine.
Se rimane la significatività statistica tra mschi e femine vuol dire che c’è qualcosa che non va nell’analisi o nei campioni di dati. \
Esempio:\
Ci chiediamo se effettivamente l’altezza dipenda dal genere ( maschio / femmina ).
Eseguiamo quindi uno “scambio” di gruppi di dati, se la significatività rimane allora c’è qualcosa che non va, mentre se la significatività non rimane allora possiamo accettare l’ipotesi.

Ci si chiede: Quando un’osservazione è significativa?

Nella costruzione di un classificatore per distinguere tra due classi, è utile sapere se le variabili di input mostrano una reale differenza tra le classi.
Il T-test valuta se le medie della popolazione di due campioni sono diverse.

Supponiamo di misurare il QI di 20 uomini e 20 donne. Un gruppo è in media più intelligente dell’altro? 

- **Ipotesi nulla (H0):** Le medie dei due gruppi sono uguali

- **Ipotesi alternativa (Ha):**  Le medie dei due gruppi non sono uguali

Se t > v, allora l'osservazione è significativa per il livello α.

1. I due array contenenti i QI dei due gruppi ( maschi,femmine) vengono normalizzati (parametri: media, dev_std, cardinalità )
2. Viene usata la libreria scipy.stats.ttest_ind per recuperare il valore di t e di p 
3. Dato il numero di gradi di libertà viene ritornato il valore critico per il test a due code
4. Se t > v l'osservazione è significativa per il livello α 

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

dataset_QI_maschi = [119, 126, 130, 121, 121, 117, 139, 130, 120, 114, 120]
dataset_QI_femmine = [130, 113, 111, 117, 122, 118, 120, 114, 112, 116, 120]

# Esegue il t-test
t_statistic, p_value = stats.ttest_ind(dataset_QI_maschi, dataset_QI_femmine)

# Genera un campione di statistiche t
np.random.seed(0)
n_simulations = 10000
simulated_t = np.random.standard_t(len(dataset_QI_maschi) + len(dataset_QI_femmine) - 2, size=n_simulations)

# Crea un grafico a densità per le statistiche t
plt.figure(figsize=(8, 5))
plt.hist(simulated_t, bins=30, density=True, alpha=0.6, label='Distribuzione t simulata')
plt.axvline(t_statistic, color='red', linestyle='--', linewidth=2, label='Statistiche t osservata')
plt.xlabel('Statistiche t')
plt.ylabel('Densità')
plt.title('Distribuzione delle Statistiche t')
plt.legend()

# Stampa il risultato del t-test
print("Statistiche t:", t_statistic)
print("Valore p:", p_value)

# Valuta la significatività del p-value
if p_value < 0.05:
    print("Il valore p è significativo (p < 0.05)")
else:
    print("Il valore p non è significativo (p >= 0.05)")

# Mostra il grafico
plt.show()
````

Esegue il test di Kolmogorov-Smirnov a due campioni per la bontà dell'adattamento.

Questo test confronta le distribuzioni continue sottostanti F(x) e G(x) di due campioni indipendenti.

Supponiamo di voler verificare l'ipotesi nulla che due campioni siano stati estratti dalla stessa distribuzione. Scegliamo un livello di confidenza del 95%; ovvero, rifiuteremo l'ipotesi nulla a favore dell'alternativa se il valore p è inferiore a 0,05.

````python
import scipy.stats as stats

dataset_QI_maschi = [119, 126, 130, 121, 121, 117, 139, 130, 120, 114, 120]
dataset_QI_femmine = [130, 113, 111, 117, 122, 118, 120, 114, 112, 116, 120]

# Esegue il KS test
ks_statistic, p_value = stats.ks_2samp(dataset_QI_maschi, dataset_QI_femmine, alternative="two-sided")

# Stampa il risultato del KS test
print("Statistiche KS:", ks_statistic)
print("Valore p:", p_value)

# Valuta la significatività del p-value
if p_value < 0.05:
    print("Il valore p è significativo (p < 0.05)")
else:
    print("Il valore p non è significativo (p >= 0.05)")

````

Soluzione: 
1. abbiamo usato la libreria di Python np.random.permutation che, prendendo in ingresso un array di elementi ne restituisce una permutazione casuale di questi.

2. E' stato inizializzato un dizionario contenente le possibili permutazioni dell'array ( chiavi )

3. Abbiamo eseguito 1000000 cicli di permutazioni casuali usando np.random.permutation che genera una permutazione dell'array in ingresso,e viene incrementato il contatore associato nel dizionario

4. Abbiamo costruito un istogramma in cui vengono graficate sull'asse delle ascisse le possibili permutazioni mentre sulle ordinate le relative frequenze con cui sono state estratte

````python
from itertools import permutations
import numpy as np
import matplotlib.pyplot as plt
a = [1,2,3,4]

#Non ci sono le virgole nelle chiavi del dizionario perche np.random.permutation(array) ritorna una lista di elementi che usa lo spazio come separatore 
#Tutte le permutazioni sono state messe a 0 ( frequenza di estrazione sequenza )
D={'[4 2 3 1]' : 0, '[3 2 4 1]' : 0, '[1 4 2 3]' : 0,
   '[4 3 1 2]' : 0, '[2 4 3 1]' : 0, '[1 2 4 3]' : 0, 
   '[4 2 1 3]' : 0, '[1 2 3 4]' : 0, '[4 3 2 1]' : 0,
   '[3 2 1 4]':  0, '[2 3 4 1]':  0, '[2 1 4 3]':  0,
   '[1 3 4 2]':  0, '[3 4 2 1]':  0, '[4 1 3 2]':  0, 
   '[3 4 1 2]':  0, '[3 1 2 4]':  0, '[2 3 1 4]':  0, 
   '[2 4 1 3]':  0, '[4 1 2 3]':  0, '[2 1 3 4]':  0, 
   '[1 3 2 4]':  0, '[3 1 4 2]':  0, '[1 4 3 2]':  0 }

array = np.array(a)

#Esegue 1 milione di permutazioni casuali dell'array a = [1,2,3,4]
for i in range(0,999999):
  shuffled = np.random.permutation(array)
  #La corrispondente chiave nel dizionario incrementa il contatore
  D[str(shuffled)] += 1

plt.bar(range(len(D)), list(D.values()), align='center')
plt.xticks(range(len(D)), list(D.keys()), rotation="vertical")

plt.title('Test di permutazione uniforme')
plt.xlabel('Permutazioni')
plt.ylabel('Frequenza')

plt.show()
````

**Conclusione**: Si può oservare che tutte le possibili permutazioni hanno la stessa probabilità di occorrenza. Il grafico delle frequenze delle permutazioni casuali mostra una distribuzione uniforme, indicando che non vi è alcuna preferenza o tendenza verso una specifica disposizione dei numeri nell'array.

Questo risultato conferma l'aspettativa teorica della permutazione uniforme, in cui tutte le permutazioni sono equiprobabili.

La **curva ROC** (Receiver Operating Characteristic) e l'**AUC** (Area Under the Curve) sono metriche utilizzate per valutare le prestazioni di un modello di classificazione binaria. Queste metriche sono ampiamente utilizzate in diversi settori, come l'apprendimento automatico, la medicina e l'ingegneria.  
  >La Curva ROC rappresenta graficamente la relazione tra il tasso di veri positivi (TPR) e il tasso di falsi positivi(FPR), al variare della soglia di classificazione del modello. Per calcolarla viene valutato il modello a diverse soglie di classificazione dove ogni soglia determina un punto sulla curva.  

  >L'AUC è la misura dell'area sottesa dalla curva ROC, e fornisce un singolo valore numerico che indica la capacità del modello di classificazione di discriminare correttamente le classi positive e negative.  
      - se valore di AUC=1 rappresenta una classificazione perfetta.  
      - se valore di AUC=0.5 indica una classificazione casuale.  

In altre parole maggiore è l'AUC, migliore è la capacità predittiva del modello, ovvero  Un AUC=0.5 indica che il modello non è in grado di distinguere tra le classi, mentre un valore prossimo a 1 indica un'elevata capacità di discriminazione.


Soluzione:

- Viene usata la libreria sklearn.metrics.roc_curve che prende in ingresso un array contenente i valori reali e i valori predetti dal modello classificatore e ritorna fpr,tpr

- Viene usata la libreria sklearn.metrics.auc che ritorna il valore di area under curve

1. Viene caricato un array contenente le etichette "reali" quindi 1 = covid positivo mentre -1 vuol dire covid negativo

2. Viene caricato un array contenente le etichette "predette dal modello" quindi 1 = covid positivo mentre -1 vuol dire covid negativo

3. Con sklearn.metrics.roc_curve viene ritornato: false positive rate, true positive rate e threshold

4. Vengono graficati tpr,fpr ottenuti dal punto 3. e anche il valore di AUC calcolato

````python
import matplotlib.pyplot as plt
import sklearn.metrics as metrics

#Inizializzo i valori reali con una lista vuota
y_true=[]

#Inizializzo i valori predetti con una lista vuota
preds=[]

#Reali: Ci sono 541 negativi al covid
for i in range(0,541):
  y_true.append(-1)
#Reali: Ci sono 359 positivi al covid
for i in range(0,359):
  y_true.append(1)

#Predetti: Ci sono 631 negativi al covid
for i in range(0,631):
  preds.append(-1)
#Predetti: Ci sono 269 negativi al covid
for i in range(0,269):
  preds.append(1)

#Calcolo fpr, tpr, threshold con la funzione libreria sklearn.metrics.roc_curve
fpr, tpr, threshold = metrics.roc_curve(y_true, preds)
#Calcolo auc con funzione sklearn.metrics.auc
roc_auc = metrics.auc(fpr, tpr)

#Rappresentazione dei dati
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')

#Limitata la visualizzazione dell'asse ascisse con un valore compreso tra 0 e 1
plt.xlim([0, 1])
#Limitata la visualizzazione dell'asse ordinate con un valore compreso tra 0 e 1
plt.ylim([0, 1])
#Etichetta asse ordinate
plt.ylabel('True Positive Rate')
#Etichetta asse ascisse
plt.xlabel('False Positive Rate')
plt.show()
````

**Conclusione**: Il modello presenta un Area Under Curve pari a 0.87, se consideriamo che un valore di AUC pari a 1 presenta un modello classificatore che non "sbaglia mai", un valore di 0.87 è da considerarsi buono, quindi generalmente il classificatore è "affidabile" nelle previsioni che esegue.

##Task 1

di regressione (lineare o non lineare) su un vostro

**Soluzione**:
- Viene studiato un dataset che mette a confronto il numero di morti con i casi di malattia del covìvid-19.

Per eseguire uno studio completo è necessario:
1. Rimuovere gli Outlier.
2. Normalizzare i dati.
3. Unione variabili altamente correlate in singola variabile.

1.1 Viene usata la funzione di libreria Python  sklearn.linear_model.LinearRegression che permette di stimare una retta di regressione che sia aderente (fit) ai dati sottposti in ingresso

1.2 Viene recuperato il valore di r^2, cioè coeff di determinazione che risulta essere 0,8 circa .. quindi in questo caso la correlazione tra le due variabili è decisamente forte

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from scipy import stats

# Carica il dataset da gitHub
url = 'https://github.com/mattiadolce/BigData/blob/main/datasetCovid2021.csv?raw=true'
dataset = pd.read_csv(url, encoding='utf-8')


# Definizione dell'array x, contenente i valori delle variabili indipendenti.
# Utilizziamo np.array per creare un array numpy, e reshape((-1, 1)) per trasformarlo in una matrice bidimensionale con una colonna.
# Definizione dell'array y, contenente i valori delle variabili dipendenti con degli outlier.
# x = np.array([0,1,2,2,2,3,4,4,5,5,3,2,4,3,4,3,5,5,2,0,1,1,2,2,2,1]).reshape((-1, 1))
# y = np.array([500,550,600,620,595,610,722,700,1000,900,650,650,850,800,890,750,950,900,600,500,550,600,690,600,5000,5500])
x = dataset['cases'].values.reshape((-1, 1))
y = dataset['deaths'].values


# Rilevamento degli outlayer utilizzando il test z-score
z_scores = np.abs(stats.zscore(y))
threshold = 3  # Soglia di rilevamento degli outlayer
outliers = np.where(z_scores > threshold)

# Creazione del grafico per visualizzare il dataset con gli outlier
plt.scatter(x, y, color='blue', label='Data')
plt.scatter(x[outliers], y[outliers], color='red', label='Outliers')
plt.xlabel('Casi')
plt.ylabel('Morti')
plt.title('Dataset con outlier')
plt.legend()
plt.show()

#creazione degli array con gli outlayer rimossi
x_cleaned = np.delete(x, outliers).reshape((-1, 1))
y_cleaned = np.delete(y, outliers)

# Suddivisione del dataset in set di addestramento e di test
# Utilizziamo la funzione train_test_split per dividere il dataset in un set di addestramento e un set di test.
# Specifichiamo test_size=0.2 per indicare che vogliamo il 20% dei dati per il test, e random_state=42 per riproducibilità.
x_train, x_test, y_train, y_test = train_test_split(x_cleaned, y_cleaned, test_size=0.2, random_state=42)

# Creiamo un'istanza del modello di regressione lineare utilizzando la classe LinearRegression fornita dalla libreria scikit-learn.
model = LinearRegression()

# Addestriamo il modello chiamando il metodo fit() sull'istanza del modello.
# Passiamo i dati di addestramento x_train e y_train per addestrare il modello sulla relazione tra le variabili indipendenti e dipendenti.
model.fit(x_train, y_train)

# Utilizziamo il metodo predict() sull'istanza del modello per ottenere le previsioni per i dati di addestramento e di test.
# Passiamo x_train per ottenere le previsioni corrispondenti alle variabili indipendenti di addestramento,
# e x_test per ottenere le previsioni corrispondenti alle variabili indipendenti di test.
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

# Utilizziamo diverse metriche per valutare le performance del modello.
# mean_squared_error calcola l'errore quadratico medio tra i valori osservati e le previsioni per entrambi i set di addestramento e di test.
# r2_score calcola il coefficiente di determinazione (R-squared) tra i valori osservati e le previsioni per entrambi i set di addestramento e di test.
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

# Stampa delle metriche di valutazione
print()
print("Scarto quadratico medio (train):", mse_train)
print("Scarto quadratico medio (test):", mse_test)
print("Coefficente di determinazione (R quadro) (train):", r2_train)
print("Coefficente di determinazione (R quadro) (test):", r2_test)
print()

# Creazione del grafico per visualizzare i risultati del modello
plt.scatter(x_train, y_train, color='blue', label='Train Data')  # Disegna i punti del set di addestramento
plt.scatter(x_test, y_test, color='green', label='Test Data')  # Disegna i punti del set di test
plt.plot(x_train, y_train_pred, color='red', label='Regression Line')  # Disegna la linea di regressione
plt.xlabel('Casi')
plt.ylabel('Morti')
plt.title('Regressione lineare')
plt.legend()
plt.show()

# Calcolo dell'errore residuo
residuals_train = y_train - y_train_pred
residuals_test = y_test - y_test_pred

# Creazione del grafico per visualizzare l'errore residuo
print()
plt.scatter(y_train_pred, residuals_train, color='blue', label='Errore residuo (train)')
plt.scatter(y_test_pred, residuals_test, color='green', label='Errore residuo (test)')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Valori previsti')
plt.ylabel('Errore residuo')
plt.title('Grafico errore residuo')
plt.legend()
plt.show()
````

**Conclusione**: Dopo aver rimosso l'outlier dal dataset dei decessi, abbiamo applicato con successo la regressione lineare per analizzare la relazione tra il numero di casi di COVID-19 e il numero di morti. Il grafico della regressione lineare ha evidenziato una tendenza positiva, mentre il grafico dell'errore residuo ha confermato l'accuratezza del modello nella previsione dei decessi. Questi risultati suggeriscono che il numero di casi influisce direttamente sul numero di morti, fornendo informazioni preziose per la comprensione dell'impatto della pandemia.

##Task 2

**Consegna**: Selezionare un dataset opportuno all’uso del clustering agglomerativo, usare le librerie Python per crearne l’albero relativo, calcolarne il suo dendogramma e graficarlo opportunamente.

Il clustering agglomerativo è un algoritmo di apprendimento utilizzato per raggruppare insieme gli oggetti simili 
in base alla loro distanza o somiglianza.Il processo di clustering agglomerativo inizia assegnando a ciascun oggetto un cluster separato. 
Successivamente, viene calcolata una matrice di distanza o di somiglianza tra gli oggetti, che può essere basata su diverse metriche.
In fine, i cluster più simili vengono uniti per formare cluster più grandi. Si ripete questa operazione iterativamente fino a quando tutti 
gli oggetti non sono stati raggruppati in un unico cluster, o fino a quando viene soddisfatto un criterio di stop definito a priori, come il numero di cluster desiderato.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Caricamento del dataset Iris
iris = load_iris()
data = iris.data

# Calcolo della matrice di distanza utilizzando la metrica di euclide
dist_matrix = linkage(data,metric='euclidean', method='ward')

# Creazione del dendrogramma
plt.figure(figsize=(10, 6))
dendrogram(dist_matrix, labels=iris.target)
plt.title("Dendogramma - Dataset Iris")
plt.xticks(fontsize=7,rotation = 90)
plt.xlabel("Campioni")
plt.ylabel("Distanza")
plt.show()

# Grafico iniziale del dataset
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=iris.target)
plt.title("Grafico iniziale del dataset Iris")
plt.xlabel("Prima caratteristica")
plt.ylabel("Seconda caratteristica")
plt.show()

cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')  
labels = cluster.fit_predict(data)

# Grafico delle varie cluster
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='rainbow')
plt.title("Grafico delle varie cluster - Dataset Iris")
plt.xlabel("Prima caratteristica")
plt.ylabel("Seconda caratteristica")
plt.show()
````

**Conclusione**: Utilizzando il dendrogramma e l'algoritmo di clustering gerarchico, siamo stati in grado di identificare la struttura dei raggruppamenti nel dataset "iris" e visualizzare la distribuzione dei dati in base ai cluster.
